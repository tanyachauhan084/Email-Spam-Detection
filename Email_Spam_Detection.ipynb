```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import recall_score
```

```python
email_data= pd.read_csv("mail_data.csv")
email_data
```

```python
email_data.shape
```

```python
email_data['Category'].value_counts()
```

```python
email_data.isnull().sum()
```

```python
email_data['Category'].value_counts()
```

```python
encoder = LabelEncoder()
encoder
```

```python
email_data['Category']= encoder.fit_transform(email_data['Category'])
```

```python
email_data
```

```python
X= email_data['Message']
X
```

```python
Y= email_data['Category']
Y
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size= 0.2,  random_state= 0)
```

```python
feature_extraction = TfidfVectorizer(min_df=1, stop_words= 'english', lowercase= True)
X_train =feature_extraction.fit_transform(X_train)
X_test= feature_extraction.transform(X_test)
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
print(X_test)
```

```python
print(Y_test)
```

```python
model= LogisticRegression()
model
```

```python
model.fit(X_train, Y_train)
```

```python
predict_X_train= model.predict(X_train)
recall_X_train= recall_score(predict_X_train, Y_train)
recall_X_train*100
```

```python
predict_X_test= model.predict(X_test)
recall_X_test= recall_score(predict_X_test, Y_test)
recall_X_test*100
```
